```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt




# Set the style for the plots
plt.style.use('ggplot')

# Plot CGPA values for each student
plt.figure(figsize=(10, 6))
plt.plot(df.index, df['CGPA'], marker='o', linestyle='-', label='CGPA')

plt.title('CGPA of Each Student')
plt.xlabel('Student Index')
plt.ylabel('CGPA')
plt.legend(title='GPA Type')
plt.show()

print('done')
```

```python
# Set the style for the plots
plt.style.use('ggplot')

# Plot histogram for CGPA values
plt.figure(figsize=(14, 8))
plt.hist(df['CGPA'], bins=20, edgecolor='black')

plt.title('Histogram of CGPA')
plt.xlabel('CGPA')
plt.ylabel('Frequency')
plt.show()

print('done')
```

```python

# Visualizing CGPA with interactive tooltips using Plotly

import plotly.express as px

# Load the data


# Plotting CGPA with interactive tooltips
fig = px.scatter(df, x='Roll No.', y='CGPA', hover_data=['Students', 'Branch', 'CGPA'], title='CGPA Visualization')

# Show the plot
fig.show()
```

```python

# Filter the data
filtered_df = df[(df['CGPA'].between(6, 9, inclusive='both')) & (df['Skill'] == 1)]

# Create a scatter plot
plt.figure(figsize=(12, 8))
sns.scatterplot(data=filtered_df, x='CGPA', y='Roll No.', hue='Branch', palette='deep')
plt.title('Students with CGPA between 6 and 9 and Skill marked as Yes')
plt.xlabel('CGPA')
plt.ylabel('Roll Number')
plt.legend(title='Branch', bbox_to_anchor=(1.05, 1), loc='upper left')
plt.tight_layout()
plt.show()

print(f'Number of students meeting the criteria: {len(filtered_df)}')
print('\
Summary statistics of CGPA for these students:')
print(filtered_df['CGPA'].describe())

print('\
First few rows of the filtered data:')
print(filtered_df[['Students', 'Roll No.', 'Branch', 'CGPA', 'Skill']].head().to_string(index=False))
```

```python


# Convert 'Skill' and 'Placement' columns to boolean
df['Skill'] = df['Skill'].map({'Yes': True, 'No': False, 1: True, 0: False})
df['Placement'] = df['Placement'].map({'Yes': True, 'No': False, 1: True, 0: False})

# 1. Graph for students with CGPA between 6 and 9
plt.figure(figsize=(12, 6))
sns.histplot(data=df[df['CGPA'].between(6, 9)], x='CGPA', bins=30, kde=True)
plt.title('Distribution of CGPA for Students with CGPA between 6 and 9')
plt.xlabel('CGPA')
plt.ylabel('Count')
plt.savefig('cgpa_distribution.png')
```

```python
# 2. Graph for students with Skill marked as Yes
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['Skill'] == True], x='Branch')
plt.title('Distribution of Students with Skill Marked as Yes by Branch')
plt.xlabel('Branch')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('skill_distribution.png')
```

```python
# 3. Graph for students with Placement marked as Yes
plt.figure(figsize=(12, 6))
sns.countplot(data=df[df['Placement'] == True], x='Branch')
plt.title('Distribution of Students with Placement Marked as Yes by Branch')
plt.xlabel('Branch')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.savefig('placement_distribution.png')


print('Graphs have been generated and saved as PNG files.')
print('\
Summary statistics:')
print(f'Total number of students: {len(df)}')
print(f'Students with CGPA between 6 and 9: {len(df[df["CGPA"].between(6, 9)])}')
print(f'Students with Skill marked as Yes: {df["Skill"].sum()}')
print(f'Students with Placement marked as Yes: {df["Placement"].sum()}')

print('\
First few rows of the dataset:')
print(df[['Students', 'Roll No.', 'Branch', 'CGPA', 'Skill', 'Placement']].head().to_string(index=False))
```
